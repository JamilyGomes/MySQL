```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/kaggle/input/chocolate-sales/Chocolate Sales.csv")

df['Amount'] = df['Amount'].replace({'\$': '', ',': ''}, regex=True).astype(float)
df['Date'] = pd.to_datetime(df['Date'])

df = df.sample(5, random_state=42)  # Seleciona aleatoriamente 5 produtos

plt.figure(figsize=(12, 8))
plt.bar(df['Product'], df['Amount'], label='Preço', alpha=0.7, color='orange')


plt.title('Preço Total e Caixas Enviadas por Produto')
plt.xlabel('Produto')
plt.ylabel('Valor')
plt.xticks(rotation=90)
plt.show()
```
